# Práctica Día 13 — Tu primer agente con Claude Code

Esta práctica no se ejecuta en el notebook: el alumno trabaja en la terminal. El notebook sirve de guía y de cuaderno de bitácora.

## Pasos

1. Instala `claude` (o `opencode + Ollama`).
2. Genera tu `CLAUDE.md`:
   ```bash
   python3 practica_13.py --out CLAUDE.md
   ```
3. Edita el CLAUDE.md para tu proyecto real (incluso uno pequeño).
4. Lanza Claude Code en la carpeta del proyecto.
5. Pide al agente que produzca un README.md completo.
6. Pega aquí abajo la transcripción de los 4-6 turnos clave.
