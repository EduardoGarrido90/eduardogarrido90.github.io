# Práctica Día 06 — Forecast LSTM sobre demanda sintética

Objetivo: usar una LSTM para predecir las próximas 4 semanas de demanda a partir de un histórico de 24.

```python
import sys; sys.path.insert(0, '.')
from practica_06 import make_series, windows, Forecaster
import torch, torch.nn as nn, matplotlib.pyplot as plt
y = make_series(); plt.plot(y); plt.title('Demanda semanal sintética'); plt.show()
```

```python
mean, std = y.mean(), y.std()
X, T = windows((y - mean)/std, lookback=24, horizon=4)
X.shape, T.shape
```

```python
n = int(0.8*len(X))
Xt, Yt = torch.tensor(X[:n]), torch.tensor(T[:n])
Xv, Yv = torch.tensor(X[n:]), torch.tensor(T[n:])
m = Forecaster(horizon=4)
opt = torch.optim.AdamW(m.parameters(), lr=1e-3, weight_decay=1e-4)
fn = nn.MSELoss()
for ep in range(40):
    opt.zero_grad(); loss = fn(m(Xt), Yt); loss.backward(); opt.step()
m.eval()
with torch.no_grad():
    pred = m(Xv).numpy() * std + mean
    true = Yv.numpy() * std + mean
mae = abs(pred - true).mean(); mae
```
