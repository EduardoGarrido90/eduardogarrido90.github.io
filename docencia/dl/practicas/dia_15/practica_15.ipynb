```python
import sys; sys.path.insert(0, '.')
from practica_15 import chunk_text
from sentence_transformers import SentenceTransformer
import numpy as np
m = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
# Sustituye estos textos por tus propios PDFs/Markdown.
docs = {
    'circ_001.md': 'La política de teletrabajo permite 3 días remotos.',
    'tarifas.md': 'Las comisiones de mantenimiento son 0,5 % anual.',
    'tipo_fijo.md': 'La hipoteca fija incluye comisión apertura 0,5 %.',
}
chunks, sources = [], []
for s, t in docs.items():
    for c in chunk_text(t):
        chunks.append(c); sources.append(s)
embs = m.encode(chunks, normalize_embeddings=True)
embs.shape
```

```python
q = m.encode('comisiones hipoteca fija', normalize_embeddings=True)
sims = embs @ q
for i in np.argsort(-sims)[:3]:
    print(f'{sims[i]:+.3f}  {sources[i]}: {chunks[i][:100]}')
```
