```python
from sentence_transformers import SentenceTransformer
import numpy as np
m = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
docs = [
    'La hipoteca a tipo fijo incluye comisión de apertura del 0,5 %.',
    'El préstamo personal tiene un tipo de interés del 7 % TAE.',
    'El banco del parque es de madera y tiene 80 años.',
    'La sucursal abre de 8:30 a 14:00 de lunes a viernes.']
embs = m.encode(docs, normalize_embeddings=True)
q = m.encode('costes de una hipoteca', normalize_embeddings=True)
sims = embs @ q
for i in np.argsort(-sims):
    print(f'{sims[i]:+.3f}  {docs[i]}')
```
