```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import LoraConfig, get_peft_model
import torch
base = 'meta-llama/Llama-3.2-1B-Instruct'
tok = AutoTokenizer.from_pretrained(base)
model = AutoModelForCausalLM.from_pretrained(base, torch_dtype=torch.float16, device_map='auto')
model = get_peft_model(model, LoraConfig(r=8, lora_alpha=16, target_modules=['q_proj','v_proj'], lora_dropout=0.05, task_type='CAUSAL_LM'))
model.print_trainable_parameters()
```

```python
from datasets import Dataset
import random; random.seed(0)
intents = ['hipoteca', 'préstamo', 'tarjeta', 'cuenta']
data = [(f'¿Qué requisitos pide la {x}?', f'Para la {x} se requiere ingresos mínimos, antigüedad y nómina domiciliada.') for x in random.choices(intents, k=500)]
ds = Dataset.from_dict({'prompt':[a for a,_ in data], 'response':[b for _,b in data]})
ds[:3]
```
