# Práctica Día 07 — Atención a mano, verificada contra PyTorch

Objetivo: implementar el bloque de atención escalada con numpy y demostrar que coincide bit-a-bit con `torch.nn.functional.scaled_dot_product_attention`.

```python
import sys; sys.path.insert(0, '.')
from practica_07 import attention_numpy
import numpy as np, torch, torch.nn.functional as F
rng = np.random.default_rng(0)
T, d, dv = 5, 8, 4
Q, K, V = (rng.normal(size=s).astype(np.float32) for s in [(T,d),(T,d),(T,dv)])
out_np = attention_numpy(Q, K, V)
out_pt = F.scaled_dot_product_attention(*(torch.tensor(x).unsqueeze(0) for x in (Q,K,V))).squeeze(0).numpy()
abs(out_np - out_pt).max()
```
